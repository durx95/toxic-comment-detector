### Toxic comment detection using naive bayes classifition.


```python
# import libaries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score ,classification_report
```

```python
df = pd.read_csv("train.csv")

print("sample data",df.head())
print("total rows:",len(df))
```

```python
# use only one label for simlicity.
df = df[["comment_text","toxic"]]
print(df)
```

```python
# data preprocessing.

df ['comment_text'] = df["comment_text"].str.lower()
df["comment_text"] = df["comment_text"].str.replace(r'[^a-z A-Z]',' ',regex=True)
```

```python
# feature engineering

vectorizer = TfidfVectorizer(max_features=5000,stop_words = "english")

x = vectorizer.fit_transform(df["comment_text"])
y = df["toxic"]

```

```python
# train test split 
x_train, x_test , y_train , y_test = train_test_split(x,y ,test_size=0.2,random_state=42)

# modeling
model = MultinomialNB()
model.fit(x_train,y_train)

# prediction 
y_pred = model.predict(x_test)
```

```python
# model evaluation
print("accuracy:",accuracy_score(y_test,y_pred))
print("classifition_report:",classification_report(y_test,y_pred))
```

```python
# test on costom input

def predict_toxic(comment):
    comment = comment.lower()
    comment = vectorizer.transform([comment])
    result = model.predict(comment)

    if result[0]== 1:
        return "toxic comment!"
    else:
        return "clean comment"
    
print(predict_toxic("  this is good!"))    
```

```python
# save model and vectorizer
import pickle 

pickle.dump(model,open("model.pkl","wb"))

pickle.dump(vectorizer,open("vectorizer.pkl","wb"))
```
